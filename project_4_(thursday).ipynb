```python
# Get unique genres in the "Movie Genre" column
genres = df['Movie Genre'].unique()

# Print the unique genres
for genre in genres:
    print(genre)
```

```python
# split movie genre on the comma
unique_genres=set(''.join(df['Movie Genre']).split(','))
unique_genres=list(unique_genres)
unique_genres
```

```python
good_genres=[genre for genre in unique_genres if not '-' in genre]
good_genres
```

```python
bad_genres=[genre.replace('-F', '-f').replace('-N', '-n') for genre in unique_genres if '-' in genre]
bad_genres
```

```python
unique_genres=good_genres+bad_genres
unique_genres
```

```python
unique_genres
```

```python
import re

container=[]

for each_genre in unique_genres: 
  ini_str=each_genre
  res_list = [s for s in re.split("([A-Z][^A-Z]*)", ini_str) if s]
  container=container+res_list
```

```python
container
```

```python
tiny_genres=list(set(container))
tiny_genres.remove('Sci-fi')
tiny_genres.remove('Film-noir')
tiny_genres.append('Sci-Fi')
tiny_genres.append('Film-Noir')
tiny_genres
```

```python
big_list=[]
for idx, genre_long_str in df['Movie Genre'].iteritems():
  encoded=[]
  for each_genre in tiny_genres: 
    if each_genre in genre_long_str: 
      encoded.append(1)
    else: 
      encoded.append(0)
  big_list.append(encoded)
```

```python
df['Movie Genre'].head()
```

```python
genre_df=pd.DataFrame(big_list, columns=tiny_genres)
genre_df.head()
```
