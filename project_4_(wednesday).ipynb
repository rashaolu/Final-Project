```python
# import dependencies 
import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt 
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
oscars_df = pd.read_csv("oscars_df.csv")
# Review the DataFrame
pd.set_option('display.max_columns', None)
oscars_df
```

```python
# drop unwanted columns
new_df=oscars_df.drop(columns=['Movie Info', 'Genres', 'Critic Consensus', 'Film ID', 'Audience Status', 'Tomatometer Status', 'Authors', 'Streaming Release Date', 'Original Release Date', 'Oscar Year', 'Tomatometer Rating','Tomatometer Count','Audience Rating','Audience Count','Tomatometer Top Critics Count','Tomatometer Fresh Critics Count','Tomatometer Rotten Critics Count'], axis=1)
new_df.tail(20)
```

```python
# fill na with "unknown"
final_df=new_df.fillna("unknown")
final_df.head(20)
```

```python
# Count the number of null values in each column
null_counts = final_df.isnull().sum()
# Display the results
print(null_counts)
```

```python
# convert DataFrame to a csv file
# final_df.to_csv("cleaned_oscar_df.csv", index=False)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df=pd.get_dummies(final_df['Award'], dtype=float)
dummies_df.head()
```

```python
# drop nominee column
is_winner=dummies_df.drop(["Nominee"], axis=1)
is_winner
```

```python
# Concatenate the original dataset with the dummy variables
df = pd.concat([final_df, is_winner], axis=1)
# Remove the original column if desired
df = df.drop('Award', axis=1)

df
```

```python
# spelling correction
df['Movie Genre'] = df['Movie Genre'].replace('Comdey', 'Comedy')
```

```python
# Get unique genres in the "Movie Genre" column
genres = df['Movie Genre'].unique()

# Print the unique genres
for genre in genres:
    print(genre)
```

```python
# split movie genre on the comma
unique_genres=set(''.join(df['Movie Genre']).split(','))
unique_genres=list(unique_genres)
unique_genres
```

```python
good_genres=[genre for genre in unique_genres if not '-' in genre]
good_genres
```

```python
bad_genres=[genre.replace('-F', '-f').replace('-N', '-n') for genre in unique_genres if '-' in genre]
bad_genres
```

```python
unique_genres=good_genres+bad_genres
unique_genres
```

```python
unique_genres
```

```python
import re

container=[]

for each_genre in unique_genres: 
  ini_str=each_genre
  res_list = [s for s in re.split("([A-Z][^A-Z]*)", ini_str) if s]
  container=container+res_list
```

```python
container
```

```python
tiny_genres=list(set(container))
tiny_genres.remove('Sci-fi')
tiny_genres.remove('Film-noir')
tiny_genres.append('Sci-Fi')
tiny_genres.append('Film-Noir')
tiny_genres
```

```python
big_list=[]
for idx, genre_long_str in df['Movie Genre'].iteritems():
  encoded=[]
  for each_genre in tiny_genres: 
    if each_genre in genre_long_str: 
      encoded.append(1)
    else: 
      encoded.append(0)
  big_list.append(encoded)
```

```python
df['Movie Genre'].head()
```

```python
genre_df=pd.DataFrame(big_list, columns=tiny_genres)
genre_df.head()
```
